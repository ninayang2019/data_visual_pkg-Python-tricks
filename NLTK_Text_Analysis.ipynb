```
# Using NLTK.Sentence token to break paragraph into sentences
from nltk.tokenize import sent_tokenize
text="""I am very Angry. We are very Happy. It is raining in Seattle."""
tokenized_text=sent_tokenize(text)
print(tokenized_text)

# Using NlTK.word token to break sentences into words
from nltk.tokenize import word_tokenize
text="""I am very Angry. We are very Happy. It is raining in Seattle."""
tokenized_word=word_tokenize(text)
print(tokenized_word)

#Positive & Negative text determination
text="""I am very Angry. We are very Happy. It is raining in Seattle."""
positive_words=['awesome','good','nice','super','fun']
newword='happy'
positive_words.append(newword)
print (positive_words)

```

```
# Then count the frequency of each word-- Frenquency Distribution
from nltk.probability import FreqDist
fdist = FreqDist(tokenized_word)
print(fdist)

# Pick up a random frequency as 2, then the word "very" popped up.
fdist.most_common(2)

#Plot the frequency
import matplotlib.pyplot as plt
fdist.plot(30,cumulative=False)
plt.show()


```
