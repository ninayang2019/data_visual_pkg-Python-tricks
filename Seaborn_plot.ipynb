```
import pandas as pd
iris= pd.read_csv("https://raw.githubusercontent.com/shineeyang/machine-learning-1/master/Iris.csv")
iris.head()
```

```
#Seaborn is used here to do plot
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```
#CountPlot
ax = sns.countplot(x="Species", data=iris)
```

```
#Heatmap
fig=plt.gcf()
fig.set_size_inches(10,7)
fig=sns.heatmap(iris.corr(),annot=True,cmap='cubehelix',linewidths=1,linecolor='k',square=True,mask=False, vmin=-1, vmax=1,cbar_kws={"orientation": "vertical"},cbar=True)
```

```
# Pairplot: 
sns.pairplot(iris,hue='Species')
```
