```
import pandas as pd
agency= pd.read_csv("https://raw.githubusercontent.com/shineeyang/Machine-Learning/master/Requesters-Shini.csv")
agency.head()
```

```
import csv
from wordcloud import WordCloud
```

```
b=agency['agency_name'].fillna('')
```

```
a=str(b)
```

```
type(a)
print (a)
```

```
# Generate a word cloud image
wordcloud = WordCloud().generate(a)

# Display the generated image:
# the matplotlib way:
import matplotlib.pyplot as plt
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")

# lower max_font_size
wordcloud = WordCloud(max_font_size=180).generate(a)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()

# The pil way (if you don't have matplotlib)
image = wordcloud.to_image()
image.show()
```
